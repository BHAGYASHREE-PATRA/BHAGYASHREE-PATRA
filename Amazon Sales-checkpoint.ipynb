```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month')['Total Revenue'].sum()
```

```python
plt.figure(figsize=(10,6))
monthly_sales.plot(kind='line',marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

It can be infered that there used to be nill sale in the mid months of years noted

```python
df['Year'] = df['Order Date'].dt.year
```

```python
yearly_sales = df.groupby('Year')['Total Revenue'].sum()
```

```python
plt.figure(figsize=(10,6))
yearly_sales.plot(kind='line',marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

There was a sharp raise in the period 2011-12 in total sales

```python
df['YearMonth'] = df['Order Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('YearMonth')['Total Revenue'].sum().reset_index()
```

```python
plt.figure(figsize=(14,7))
monthly_sales.plot(kind='line',marker='o',color='g')
plt.title('Year-Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

In the period 2013-14 there was a surge in total sales
